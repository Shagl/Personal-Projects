```python
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_imdb= pd.read_csv('IMDBmovies.csv')
df_imdb
```

```python
df_imdb.drop(columns=["budget", "usa_gross_income","worlwide_gross_income", "metascore", 'date_published', 'duration','production_company','reviews_from_users','reviews_from_critics'],inplace=True)
df_imdb
```

```python
df_imdb['country']= df_imdb['country'].fillna('No data entered')
df_imdb['language']= df_imdb['language'].fillna('No data entered')
df_imdb['director']= df_imdb['director'].fillna('No data entered')
df_imdb['writer']= df_imdb['writer'].fillna('No data entered')
df_imdb['description']= df_imdb['description'].fillna('No data entered')
df_imdb['actors']= df_imdb['actors'].fillna('No data entered')
```

```python
def Movie_Details(data):
    details=[]
    for i in range(0, data.shape[0]):
        details.append(data['title'][i]+ ' '+data['genre'][i]+ ' '+data['country'][i]+ ' '+data['language'][i])
    return details  
```

```python
df_imdb['Movie_Details']= Movie_Details(df_imdb)
df_imdb
```

```python
movie= df_imdb[(df_imdb['avg_vote'] >= 6.9)]
movie= movie.sort_values(['avg_vote'], ascending= False)
movie
```

```python
movie.index = np.arange(1,len(movie)+1)
movie
```

```python
cv= CountVectorizer( max_features=3000)
cm= cv.fit_transform(movie['Movie_Details']) 
cm.shape
```

```python
cs= cosine_similarity(cm)
cs
```

```python
m_title= movie['title'][1]
m_title
```

```python
movie_id= movie[movie.title==m_title]['imdb_title_id'].index[0] 
movie_id
```

```python
movie_val= list(enumerate(cs[movie_id]))
movie_val
```

```python
sorted_movie_val= sorted(movie_val,reverse = True,key = lambda x: x[1])
sorted_movie_val= sorted_movie_val[1:] #birden sonra çünkü zaten en yüksek olan kendisi
sorted_movie_val
```

```python
ndex = movie[movie['title'] == 'Ek'].index[0]
istances= cs[ndex] 
orted_movies= sorted(list(enumerate(istances)) ,reverse = True,key = lambda x: x[1])
orted_movies
```

```python
print('RECOMMENDED MOVIES\n')
def recommend(movies):
    index = movie[movie['title'] == movies].index[0]
    distances= cs[index] 
    sorted_movies= sorted(list(enumerate(distances)) ,reverse = True,key = lambda x: x[1])
    sorted_movies
    for i in sorted_movies:
        recommended_movies=movie.iloc[i[0]].title
        print(recommended_movies)


movie_name= recommend("Forrest Gump")
```
